```python
# 여기에 공통과제 2번 코드를 작성해주세요. 주석을 꼼꼼하게 작성하여 왜 이렇게 시각화를 했는지 설명해주시면 더 좋습니다!
import random 

# Convert the zip object into a list
zipped_data = list(zip(df['carat'], df['cut'], df['color'], df['clarity'], df['price']))

# Convert the list into a dataframe for easier manipulation
cut_df = pd.DataFrame(zipped_data, columns = ['carat', 'cut', 'color', 'clarity', 'price'])

# Showing only 100 random samples of the data for EACH CARAT CATEGORY on the plot
random_samples = []

# Iterating over UNIQUE carat values in zipped data
for cut_value in cut_df['cut'].unique():
    # Selecting 100 random samples from the carat_df dataframe for each carat value
    random_samples.append(cut_df[cut_df['cut'] == cut_value].sample(100, replace = True))

    # Concatenating the 100 random samples into a single dataframe
    random_samples_df = pd.concat(random_samples)

print(random_samples_df)

# Colourmap
cmap = plt.cm.get_cmap('viridis')

# Plotting the data
fig, ax = plt.subplots()
fig.suptitle("The Price Distribution of Diamonds by Different Cut Grades and Carat Values \n (100 Random Samples for Each Cut Grade)", size=14)

# adjusting the markersize to make the plot more readable
for carat, cut, color, clarity, price in random_samples_df.values:
    ax.plot(cut, price, marker = 'o', color = cmap(carat), markersize = 3)
fig.colorbar(plt.cm.ScalarMappable(norm=Normalize(0, 5), cmap=cmap), ax=ax, label="Carat")

ax.set_xlabel("Cut Grade")
ax.set_ylabel("Price ($)")
ax.set_ylim(0, 5000)
plt.show()
```

```python
try:
    df = sns.load_dataset("StockX-Data-Contest-2019-3.csv")
except Exception as e:
    csv_path = r"StockX-Data-Contest-2019-3.csv"
    df = pd.read_csv(csv_path)
df
```

```python
# Difference between Sale Price and Retail Price

def str_to_integer(price):
    return int(price.replace(',', '')[1:])
```

```python
df['Sale Price'] = df['Sale Price'].apply(str_to_integer)
df['Retail Price'] = df['Retail Price'].apply(str_to_integer)
df
```

```python
df['Markup'] = ((df['Retail Price'] - df['Sale Price'])/df['Sale Price'])*100
df
```

```python
# Get a monthly average markup for each shoe brand
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Month and Year'] = df['Order Date'].dt.strftime('%Y-%m')

# Average Markup
average_markup = df.groupby(['Brand', 'Order Month and Year'])['Markup'].mean().reset_index()
average_markup

# Percentage change of the average markup
average_markup['Average Markup Change (%)'] = -round((average_markup.groupby('Brand')['Markup'].pct_change()*100),2)
average_markup
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Barplot
bar = sns.barplot(data=average_markup, x='Order Month and Year', y='Average Markup Change (%)', hue='Brand', ax=ax)

# Lineplot
line = sns.lineplot(data=average_markup, x='Order Month and Year', y='Average Markup Change (%)', hue='Brand', marker='o', linewidth=3, ax=ax)

# Set labels and title
ax.set_title('Percentage Change of Average Markup by Month-Year for Each Brand', fontsize=20)
ax.set_ylabel('Average Markup Change (%)', fontsize=10)
ax.set_xlabel('Order Month and Year', fontsize=10)
ax.set_xticklabels(average_markup['Order Month and Year'].unique(), rotation=45)

# Show the plot
plt.show()
```
